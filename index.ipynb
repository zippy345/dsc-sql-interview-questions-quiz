* There are three types of entities
  Which are :
  * 1. ONE TO ONE
       >  Each row in the first table is linked to exactly one row in the second table, and vice versa.
    2. ONE TO MANY
       >  A row in first table can be related to many rows in the second table, but each row in second table belongs to only one row in fisrt               table.
    3. MANY TO MANY
       >  Rows in first table can be linked to many rows in the second table, and vice versa.



## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
* There are four main type of joins in sql. They are:
   * 1. INNER JOIN
        > returns  an output with only matching columns between the two tables that were being joined.
     2. FULL/OUTER JOIN
        > returns an output with all columns from the two tables being joined. 
     3. LEFT JOIN
        > returns an output with all rows from the left table, and matched rows from the right table. If no match, fills with NULL
     4. RIGHT JOIN
        > returns an output with all rows from the right table, and matched rows from the left table. If no match, fills with NULL.
